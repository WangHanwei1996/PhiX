```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the free energy data file
data = np.loadtxt('output/free_energy.dat')

# Extract time and free energy columns
time = data[:, 0] / 1e3
free_energy = data[:, 1]
```

```python
file1 = pd.read_csv("reference/MOOSE_IA_2a.csv")
t1 = file1['time'][1:]
f1 = file1['TotalEnergy'][1:]

file2 = pd.read_csv("reference/MOOSE_FH_2a.csv")
t2 = file2['time'][1:]
f2 = file2['total_energy'][1:]
```

```python

fig, ax = plt.subplots(figsize=(8, 6))


ax.plot(t1, f1, label='Moose_IA_2a',  color='orange', linewidth=2)
ax.plot(t2, f2, label='Moose_FH_2a',  color='green', linewidth=2)

ax.plot(time, free_energy, 'o', color='blue', label='Present')

ax.set_xscale("log")
ax.set_yscale("log")
ax.set_xlabel("time", fontsize=14)
ax.set_ylabel("Free Energy", fontsize=14)
ax.set_title("Ostwald Ripening Free Energy Evolution", fontsize=16)
ax.legend(fontsize=12)
ax.grid(True, which='both', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig("output/png/free_energy_comparison.png", dpi=300)
```

```python

# Plot with log-log scale
fig, ax = plt.subplots(figsize=(8, 6))
ax.loglog(time, free_energy, 'b-', linewidth=1.5)

ax.set_xlabel('Time', fontsize=12)
ax.set_ylabel('Free Energy', fontsize=12)
ax.set_title('Free Energy vs Time (Log-Log Scale)', fontsize=14)
ax.grid(True, which='both', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
