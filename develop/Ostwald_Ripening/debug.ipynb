```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

```

```python
file = pd.read_csv("reference/pfhub_benchmark2a_digitized_points.csv")
```

```python

fig, ax = plt.subplots(figsize=(8, 6))

for curve_name, group in file.groupby("curve"):
    ax.plot(group["x_sim"], group["free_energy"], label=curve_name, marker="o", markersize=3, linestyle="-")

ax.set_xscale("log")
ax.set_yscale("log")
ax.set_xlabel("x_sim")
ax.set_ylabel("Free Energy")
ax.set_title("Free Energy vs x_sim (log-log)")
ax.legend(fontsize=8)
plt.tight_layout()
plt.show()
```
