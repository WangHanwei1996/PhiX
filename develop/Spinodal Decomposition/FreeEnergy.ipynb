```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read the free energy data file
data = np.loadtxt('output/free_energy.dat')

# Extract time and free energy columns
time = data[:, 0]
free_energy = data[:, 1]
```

```python

# Plot with log-log scale
fig, ax = plt.subplots(figsize=(8, 6))
ax.loglog(time, free_energy, 'b-', linewidth=1.5)

ax.set_xlabel('Time', fontsize=12)
ax.set_ylabel('Free Energy', fontsize=12)
ax.set_title('Free Energy vs Time (Log-Log Scale)', fontsize=14)
ax.grid(True, which='both', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
