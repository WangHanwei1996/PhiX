```python
import numpy as np
```

```python
c = 0.51882196
c_l = 0.52020979
c_r = 0.51727397
c_u = 0.51686777
c_d = 0.52046299
```

```python
lap_c = c_l + c_r + c_u + c_d - 4 * c
print(lap_c)
```

```python
kappa = 2
rho = 5
ca = 0.3
cb = 0.7
```

```python
mu_real = 2*rho*(c-ca)*(c-cb)*(2*c-ca-cb) + kappa*lap_c
print(mu_real)
```

```python
mu = -0.01587088
mu_l = -0.01711437
mu_r = -0.01448513
mu_u = -0.01423417
mu_d = -0.01723694
```

```python
M = 5
dt = 1e-3
```

```python
lap_mu = mu_l + mu_r + mu_u + mu_d - 4 * mu
print(lap_mu)
```

```python
c_new = c + dt * M * lap_mu
print(c_new)
# expect: 0.5188240281 (matches simulation step 1 output)
```
