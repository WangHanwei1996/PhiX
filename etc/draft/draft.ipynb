$$f_{chem}(c)=\rho_s(c-c_{\alpha})^2 (c_{\beta}-c)^2$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
c = np.linspace(0, 1, 100)

rho_s = 5

c_alpha = 0.3
c_beta = 0.7
```

```python
f = rho_s * (c - c_alpha) ** 2 * (c - c_beta) ** 2

plt.plot(c, f)
plt.xlabel('c')
plt.ylabel('f')
plt.title('Plot of f vs c')
plt.show()
```
